## Data Preparation for Recommendation 

```python
del data1['Unnamed: 0']
```

```python
user_user_recommendation = pd.pivot_table(data=data1,index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
user_user_recommendation.head()
```

```python
user_user_recommendation.index = data1['User.ID'].unique()
user_user_recommendation.head(10)
```

```python
user_user_recommendation.fillna(value=0,axis=0,inplace=True)
user_user_recommendation.head(10)
```

## Model Building and identifying the similar users for recommendation 

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import correlation,cosine
```

```python
user_user_recommendation.values
```

```python
## to get similarity score between 0 to 1, we can use another metrics called as cosine 
user_user_similarity = 1- pairwise_distances(X=user_user_recommendation.values,metric='cosine') 
user_user_similarity
```

```python
user_user_df = pd.DataFrame(user_user_similarity)
user_user_df.head()
```

```python
# to make the similarity score of user correlating with itself to 0 
np.fill_diagonal(user_user_similarity, 0)
```
